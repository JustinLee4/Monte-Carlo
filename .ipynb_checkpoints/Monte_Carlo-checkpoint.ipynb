```python
import csv
import os
import matplotlib.pyplot as plt
import numpy
```

```python
# Initialize empty lists
x, total_gridpoints, num_waters, current_energy, lowest_energy = [], [], [], [], []

with open("monte_carlo_output_test.csv", 'r') as f:
    reader = csv.reader(f)    
    for row in reader:
        # Append data to lists (converting to float as we go)
        x.append(float(row[0]))
        total_gridpoints.append(float(row[1]))
        num_waters.append(float(row[2]))
        current_energy.append(float(row[3]))
        lowest_energy.append(float(row[4]))
```

```python
for i, current_item in enumerate(lowest_energy[:-1]):
    next_item = lowest_energy[i + 1]
    if(current_item<next_item):
        print(i)
```

```python
plt.xlabel("# of Waters")
plt.ylabel("Count")
plt.hist(num_waters, bins=50)
plt.title("Distribution of # of Water (packmol internals)")
plt.show()
```

```python
plt.xlabel("kcal/mol")
plt.ylabel("Count")
plt.hist(current_energy, bins=50)
plt.title("Distribution of Energies (packmol internals)")
plt.show()
```

```python
fig, ax = plt.subplots() # Create a figure and an axes object

ax.plot(x,lowest_energy)
#ax.plot(x, current_energy, '--r')
plt.ylabel("kcal/mol")
plt.suptitle('Monte Carlo Energy (packmol internals)')
ax.set_title(f'Final Energy = {lowest_energy[-1]}')
plt.show()
```
