```python
import csv
import os
import matplotlib.pyplot as plt
import numpy
```

```python
x3, current_energy3, lowest_energy3 = [], [], []

with open("results/test_3_cluster_1.log", 'r') as f:
    reader = csv.reader(f)    
    for row in reader:
        if not row:
            continue
            
        x3.append(float(row[0]))
        current_energy3.append(float(row[2]))
        lowest_energy3.append(float(row[3]))

x4, current_energy4, lowest_energy4 = [], [], []

with open("results/test_4_cluster_1.log", 'r') as f:
    reader = csv.reader(f)    
    for row in reader:
        if not row:
            continue
            
        x4.append(float(row[0]))
        current_energy4.append(float(row[2]))
        lowest_energy4.append(float(row[3]))

x5, current_energy5, lowest_energy5 = [], [], []

with open("results/test_5_cluster_1.log", 'r') as f:
    reader = csv.reader(f)    
    for row in reader:
        if not row:
            continue
            
        x5.append(float(row[0]))
        current_energy5.append(float(row[2]))
        lowest_energy5.append(float(row[3]))

x6, current_energy6, lowest_energy6 = [], [], []

with open("results/test_6_cluster_1.log", 'r') as f:
    reader = csv.reader(f)    
    for row in reader:
        if not row:
            continue
            
        x6.append(float(row[0]))
        current_energy6.append(float(row[2]))
        lowest_energy6.append(float(row[3]))

x7, current_energy7, lowest_energy7 = [], [], []

with open("results/test_7_cluster_1.log", 'r') as f:
    reader = csv.reader(f)    
    for row in reader:
        if not row:
            continue
            
        x7.append(float(row[0]))
        current_energy7.append(float(row[2]))
        lowest_energy7.append(float(row[3]))

x8, current_energy8, lowest_energy8 = [], [], []

with open("results/test_8_cluster_1.log", 'r') as f:
    reader = csv.reader(f)    
    for row in reader:
        if not row:
            continue
            
        x8.append(float(row[0]))
        current_energy8.append(float(row[2]))
        lowest_energy8.append(float(row[3]))

```

```python
fig, ax = plt.subplots(2,2)

ax[0][0].plot(x3,current_energy3)
ax[0][0].set_ylabel("kcal/mol")
ax[0][0].set_title(f'Monte Carlo Energy (10^3 steps) \nLowest Energy = {lowest_energy3[-1]} \n time = 4s')

ax[0][1].plot(x4,current_energy4)
ax[0][1].set_ylabel("kcal/mol")
ax[0][1].set_title(f'Monte Carlo Energy (10^4 steps) \nLowest Energy = {lowest_energy4[-1]} \n time = 3s')

ax[1][0].plot(x5,current_energy5)
ax[1][0].set_ylabel("kcal/mol")
ax[1][0].set_title(f'Monte Carlo Energy (10^5 steps) \nLowest Energy = {lowest_energy5[-1]} \n time = 7s')

ax[1][1].plot(x6,current_energy6)
ax[1][1].set_ylabel("kcal/mol")
ax[1][1].set_title(f'Monte Carlo Energy (10^6 steps) \nLowest Energy = {lowest_energy6[-1]} \n time = 40s')

plt.tight_layout()
plt.show()
```
