```python
import csv
import os
import matplotlib.pyplot as plt
import numpy
```

```python
# Initialize empty lists
x, current_energy, lowest_energy = [], [], []

with open("results/test_2_cluster_1.log", 'r') as f:
    reader = csv.reader(f)    
    for row in reader:
        # THE MAGIC FIX: If the row is empty, skip it!
        if not row:
            continue
            
        # Append data to lists (converting to float as we go)
        x.append(float(row[0]))
        current_energy.append(float(row[1]))
        lowest_energy.append(float(row[2]))

print(f"Loaded {len(x)} steps of Monte Carlo data!")
```

```python
for i, current_item in enumerate(lowest_energy[:-1]):
    next_item = lowest_energy[i + 1]
    if(current_item<next_item):
        print(i)
```

```python
print(lowest_energy[-1])
```

```python
fig, ax = plt.subplots() # Create a figure and an axes object

ax.plot(x,lowest_energy)
#ax.plot(x, current_energy, '--r')
plt.ylabel("kcal/mol")
plt.suptitle('Monte Carlo Energy (Cluster 1)')
ax.set_title(f'Final Energy = {lowest_energy[-1]}')
plt.show()
```
